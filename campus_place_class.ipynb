## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df_col=['gender','workex','specialisation','status','degree_t','hsc_s']

for col in df_col:
    df[col] = le.fit_transform(df[col])
df.head()
```

## Dividing data into the target(y) and Independant variable(x)

```python
x=df.drop(["status"],axis=1)
x
```

```python
y=df["status"]
y
```

## Splitting the Data into Training and testing data

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=0)
```

## Creating Machine Learning Model

```python
# Using Logistic regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xtrain,ytrain)
```

```python
ypred=logreg.predict(xtest)
ypred
```

```python
ytest
```

## Evaluation of model

```python
from sklearn.metrics import confusion_matrix as cm,classification_report as cr,accuracy_score as acc
print(f"confusion matrix:-\n {cm(ytest,ypred)}")
print(f"\nAccuracy:- {acc(ytest,ypred)}")
```

```python
print(f"Classification Report:-\n {cr(ytest,ypred)}")
```
